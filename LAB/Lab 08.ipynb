Write a Python program to implement a Stack class that has the following functions

Push Item
Pop Item
Print the Items in the stack
Size of Stack
The top item in the stack
Check if the stack is empty

```python
class Stack:
    # creating a constructor 
    def __init__(self):
        #defining a empty list    
        self.stack=[] 
    # create a function for push the items in the list / stack  
    def pushitem(self,item):
        self.stack.append(item)
    # create a function for pop the items in the list / stack
    def popitem(self):   
        return self.stack.pop()
    # create a function for see all the items from stack
    def viewstack(self):  
        for item in self.stack:
            print(item)
    # create a function for get the size the stack
    def size(self):   
        print(len(self.stack))
    # create a function fro peek 
    def topstack(self):  
        print(self.stack[-1])
    # create a function to check the stack is empty or not 
    def isEmpty(self):  
        if self.stack==[]:
            print("The Stack Is empty")
        else:
            print("The Stack is Not empty")
    # create a menu function 
    def Menu (self):    
        while True:
            print("Menu Options")
            print("************************************")
            print("1 PushItem")
            print("2 PopItem")
            print("3 View Full Stack")
            print("4 Size of the stack")
            print("5 View the top Element of the stack")
            print("6 to check the stack is empty or not")
            print("7 Exit ")
            print("************************************")
            choice=int(input("Enter your choice: "))
            if choice== 1:
                item=int(input("Enter: "))
                self.pushitem(item)
            elif choice== 2:
                self.popitem()
            elif choice== 3:
                self.viewstack()
            elif choice== 4:
                self.size()
            elif choice== 5:
                self.topstack()
            elif choice== 6:
                self.isEmpty()
            elif choice== 7:
                print("Thank You")
                break
            else:
                print("Chose a Valid Output")
                
obj=Stack()
obj.Menu()
```
